```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns

num_cols, cat_cols
```

Esse código separa automaticamente as colunas do seu DataFrame em: numéricas (num_cols) e categóricas (cat_cols).

Variáveis numéricas: study_hours_week, attendance_rate, previous_scores, sleep_hours, final_grade, age Variáveis categóricas: gender, parental_education, internet_quality, health_status, etc.

```python
# Histogramas
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribuição da variável numérica: {col}")
    plt.show()

# Boxplots
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot da variável numérica: {col}")
    plt.show()
```

```python
df.describe()                 # numéricas
df.describe(include='object') # categóricas
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Gerar o describe das categóricas
desc = df.describe(include='object')

# Selecionar a linha "unique"
unique_values = desc.loc['unique']

# Plotar
plt.figure(figsize=(12,6))
unique_values.plot(kind='bar', color='orange')
plt.title("Quantidade de categorias únicas por variável categórica")
plt.ylabel("Nº de categorias diferentes")
plt.xlabel("Variáveis categóricas")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="Blues")
plt.title("Matriz de Correlação")
plt.show()
```
