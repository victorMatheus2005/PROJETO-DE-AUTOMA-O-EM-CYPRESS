Na Etapa 3 foi utilizado como modelo baseline a Regressão Linear, com divisão dos dados em 60% treino, 20% validação e 20% teste. As métricas avaliadas foram MSE, RMSE, MAE e R², além da análise de overfitting e resíduos. Este modelo serviu como referência mínima de desempenho.

Para a Etapa 4, foi escolhido o modelo Random Forest Regressor, por apresentar maior capacidade de capturar relações não lineares e permitir otimização via hiperparâmetros.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

modelo = RandomForestRegressor(random_state=42)

random_search = RandomizedSearchCV(
    estimator=modelo,
    param_distributions=param_dist,
    n_iter=20,
    cv=5,
    scoring="r2",
    n_jobs=-1,
    random_state=42
)

random_search.fit(X_train, y_train)
```

```python
best_params = random_search.best_params_
best_model = random_search.best_estimator_

best_params
```
